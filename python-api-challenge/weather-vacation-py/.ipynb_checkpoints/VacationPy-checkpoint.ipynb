```python
from citipy import citipy
import gmaps
from config import gkey
import pandas as pd
import requests
import random
import json
import matplotlib.pyplot as plt 
```

```python
weather_csv = 'Resources/weather_data.csv'

weather_df = pd.read_csv(weather_csv)

weather_df
```

```python
dropna_weather_df = weather_df.dropna()
del dropna_weather_df["Unnamed: 0"]
dropna_weather_df.head(20)
```

```python
gmaps.configure(api_key=gkey)

# Locations
locations = dropna_weather_df[["city_name", "country"]]

humidity = dropna_weather_df["humidity"].astype(float)
```

```python
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius=2)
fig.add_layer(heat_layer)

fig

if TraitError:
    print("did not capture the correct data when creating the CSV.")
```
