```python
player_dems = purchase_data_df.loc[:, ["Gender", "SN", "Age"]]
player_dems = player_dems.drop_duplicates()

num_players = player_dems.count()[0]
pd.DataFrame({"Total Players":[num_players]})
```

```python
#Calcualtions for find information needed
average_purchase = purchase_data_df["Price"].mean()
total_purchase_val = purchase_data_df["Price"].sum()                  
total_purchase = purchase_data_df["Price"].count()
item_count = len(purchase_data_df['Item ID'].unique())

#DataFrame to hold potential values
summary_table = pd.DataFrame({"Number of Unique Items": item_count,
                             "Total Revenue": [total_purchase_val],
                             "Number of Purchases": [total_purchase],
                             "Average Price": [average_purchase]})

summary_table = summary_table.round()
summary_table["Average Price"] = summary_table["Average Price"].map("${:,.2f}".format)
summary_table["Number of Purchases"] = summary_table["Number of Purchases"].map("${:,.2f}".format)
summary_table["Total Revenue"] = summary_table["Total Revenue"].map("${:,.2f}".format)
summary_table = summary_table.loc[:,["Number of Unique Items", "Average Price", "Number of Purchases", "Total Revenue"]]

#call function to diisplay the table
summary_table
```
